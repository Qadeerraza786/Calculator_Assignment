```python
import os
import random
from google import genai
from google.colab import userdata

# --- Configuration (STEP 1: Load the API Key) ---
# NOTE: Set your API key in Colab's secrets manager under the name 'GEMINI_API_KEY'
try:
    # This securely retrieves your key from Colab's secrets manager
    API_KEY = userdata.get("GEMINI_API_KEY")
    if API_KEY is None:
      raise ValueError("GEMINI_API_KEY not found in Colab secrets.")
except ValueError as e:
    print(f"FATAL ERROR: {e}")
    print("Please set the GEMINI_API_KEY in Colab secrets before running the script.")
    exit()
except Exception as e:
    print(f"An unexpected error occurred while retrieving the API key: {e}")
    exit()


# Initialize the Gemini client
client = genai.Client(api_key=API_KEY)
# Use a model that's good for multi-turn chat
chat = client.chats.create(model="gemini-2.5-flash")


# --- The Main Conversational Loop (STEP 2: Use the AI) ---
print("ðŸ¤– AI Chatbot: Hello! I can help you with real-world problems and questions. Type 'bye' to end the chat.")
print("-" * 70)

while True:
    user_input = input("ðŸ‘¤ You: ")

    # Check for the exit command
    if user_input.lower().strip() in ["bye", "exit", "quit"]:
        print("ðŸ¤– AI Chatbot: Goodbye! Feel free to ask me anything anytime. ðŸ‘‹")
        break

    try:
        # Send the user's message to the Gemini model
        # The chat object maintains the conversation history automatically!
        response = chat.send_message(user_input)

        # Print the AI's response
        print("ðŸ¤– AI Chatbot:", response.text)

    except Exception as e:
        # Handle API errors or connection issues
        fallback_responses = [
            "Oops! I ran into a communication issue. Can you try again?",
            "My connection seems a bit fuzzy right now. Please re-enter your question.",
        ]
        print("ðŸ¤– AI Chatbot:", random.choice(fallback_responses))
        # Optional: Print the exact error for debugging
        # print(f"DEBUG ERROR: {e}")
```
